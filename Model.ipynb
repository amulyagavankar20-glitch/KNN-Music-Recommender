```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
songs = pd.read_csv("Data\data.csv")
```

```python
songs.head()
```

```python
song_info = songs[["name","artists"]]
```

```python
X = songs.drop(["id","name","artists","release_date"], axis=1)
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors
```

```python
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
X["popularity"] = X["popularity"] * 2
X["danceability"] = X["danceability"] * 1.5
```

```python
knn = NearestNeighbors(n_neighbors=10, metric="cosine")
knn.fit(X_scaled)
```

```python
def recommend(song_index):
    distances, indices = knn.kneighbors([X_scaled[song_index]])
    return songs.iloc[indices[0]][["name","artists","popularity"]]
```

```python
recommend(100)
```

```python
def recommend_song(song_name):
    idx = songs[songs["name"] == song_name].index[0]
    distances, indices = knn.kneighbors([X_scaled[idx]])
    return songs.iloc[indices[0]][["name","artists"]]
```

```python
recommend_song("Blinding Lights")
```
