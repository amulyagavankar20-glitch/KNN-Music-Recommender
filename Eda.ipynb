```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
songs = pd.read_csv("Data\data.csv")
artist = pd.read_csv("Data\data_by_artist.csv")
genres = pd.read_csv("Data\data_by_genres.csv")
year = pd.read_csv("Data\data_by_year.csv")
songs_genres = pd.read_csv("Data\data_w_genres.csv")
```

```python
songs = songs.drop(["id","name","artists","release_date"], axis=1)
```

```python
plt.hist(songs["duration_ms"], bins=50)
plt.title("Song Duration Distribution")
plt.xlabel("Duration (ms)")
plt.ylabel("Frequency")
plt.show()
```

```python
popularity_counts = songs["popularity"].value_counts().head(10)
plt.pie(popularity_counts, labels=popularity_counts.index, autopct='%1.1f%%')
plt.title("Top 10 Popularity Distribution")
plt.show()
```

```python
sns.histplot(songs["energy"], bins=50)
plt.show()
```

```python
sns.boxplot(x=songs["danceability"])
plt.show()
```

```python
sns.scatterplot(x="energy", y="danceability", data=songs)
plt.show()
```

```python
fig, axes = plt.subplots(2,2, figsize=(12,8))

sns.histplot(songs["popularity"], ax=axes[0,0])
sns.histplot(songs["danceability"], ax=axes[0,1])
sns.histplot(songs["energy"], ax=axes[1,0])
sns.histplot(songs["valence"], ax=axes[1,1])

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(songs.corr(), cmap="coolwarm")
plt.show()
```
